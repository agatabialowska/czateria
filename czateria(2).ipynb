Zadanie 1
Które z następujących zadań wymagają w Twojej opinii inteligencji od człowieka:
• wypełnianie deklaracji PIT, tak

• streszczanie tekstu, tak - wymaga umiejętności pisania, czytania i myślenia logicznego, żeby zachować logiczny sens wypowiedzi, dlatego tylko człowiek może to wykonać

• tłumaczenie tekstu, tak - wymaga znajomości języka obcego, co wymaga inteligencji

• klasyfikacja tekstu do kategorii tematycznych, tak - wymaga umiejętności czytania ze zrozumieniem

• odpowiadanie na proste pytania zadawane w języku naturalnym (np. polskim), tak - wymaga umiejętności mówienia, a sama umiejętność wykorzystania mowy należy do własności wyższych form inteligencji

• układanie rozkładu jazdy transportu miejskiego, tak - wymaga myślenia

• programowanie (pisanie programów komputerowych), tak - wymaga obycia z pracą na komputerze, samej umiejętności posługiwania się komputerem, a także znajomości matematyki, logiki i języka

• „programowanie” kanałów telewizyjnych, tak - wymaga umiejętności czytania i possługiwania się telewizorem i pilotem

• testowanie oprogramowania, tak - tak samo jak w przypadku programowania komputerowego, wymagaobycia i znajomości komputera

• komponowanie muzyki, tak - ta umiejśtnośc jest przypisana istotom inteligentnym, zwierzęta nie są w stanie wydawać dźwięków które tworza melodię, tym bardziej nie umieja posługiwac się instrumentami

• rozwiązywanie układów równań, tak - wymaga umiejętności nauki matematyki

• symboliczne obliczanie pochodnych funkcji, tak - wymaga umiejętności nauki matematyki

• symboliczne całkowanie funkcji, tak - wymaga umiejętności nauki matematyki

• kierowanie samochodem. tak - wymaga koordynowania wzroku i ruchu a także maksymalnego skupienia uwagi na drodze i obserwowania sytuacji drogowych

Każde z tych zdań wymaga inteligencji od człowieka, ponieważ nawet najprostsze, z tych zadań wymaga myślenia, a każdy człowiek jest rozumny, inne istoty nie posiadaja IQ, co uniemożliwia im wykonywanie tych czynności.

Zadanie 2
Które z następujących problemów można uznać za mieszczące się w zakresie sztucznej
inteligencji:

• streszczanie tekstu, tak - czatboty potrafią analizować tekst i go przetważać

• tłumaczenie tekstu, tak - używanie translatorów internetowych pozwala na tłumaczenie tekstu

• klasyfikacja tekstu do kategorii tematycznych, tak - analiza tekstu dzięki cztbotom

• odpowiadanie na proste pytania zadawane w języku naturalnym, tak - czaty mają wgrane gotowe rozmowy i odpowiedzi dzięki czemu człowiek może prowadzić z nimi "rozmowę"

• rozwiązywanie układów równań, tak - sztuczna inteligencja ma wgraną analizę danych, dzięki czemu może wykonywać operacje na liczbach

• układanie rozkładu jazdy, tak - na podstawie różnychy danych, które posiada AI może stworzyć rozkład jazdy

• rozwiązywanie układów równań liniowych, tak - sztuczna inteligencja ma wgraną analizę danych, dzięki czemu może wykonywać operacje na liczbach

• symboliczne obliczanie pochodnych, tak - sztuczna inteligencja ma wgraną analizę danych, dzięki czemu może wykonywać operacje na liczbach

• symboliczne całkowanie, tak - sztuczna inteligencja ma wgraną analizę danych, dzięki czemu może wykonywać operacje na liczbach

• kierowanie samochodem. tak - dzięki zaawansowanym systemom technologicznym, sztuczna inteligencja przejmuje wiele funkcji w nowych autach, co może przyczynić się do większego bezpieczeństwa na drodze i zminimalizowaniu błędu ludzkiego

Zadanie 3
Które z poniższych rodzajów komunikacyjnego zachowania człowieka mogą być
obecnie skutecznie imitowane przez sztuczne systemy (odpowiednio oprogramowane
maszyny):

• rozmowa towarzyska, tak - czat może odpowiadać na proste pytania, lecz brakuje mu elementu naturalności, którą posiada człowiek

• dyskusja polityczna, nie - każdy człowiek ma ine poglądy polityczne, AI jest zazwyczaj bardzo obiektywne i nie przyjmie jednej strony sporu. Dyskusje polityczne są często zabarwione emocjonalnie, czego sztucznej inteligencji brakuje.

• dyskusja naukowa, tak - czatboty maja wgrane ogromne bazy danych, dlatego mogą przywoływać setki artykułów naukowych, dzięki czemu można się nauczyć wielu nowych rzeczy

• odpowiadanie na pytania klientów w telefonicznej infolinii, tak - AI może mieć wgraną pewna bazę pytań którą zadaja klienci, i pewną pulę możliwych odpowiedzi, mimo to i tak klient może zadać pytanie, ktorego brakuje w  bazie


Zadanie 4
Do zadań skorzystaj z załączonej listy Chatbotów (Chatterbotów) lub znajdź własne.
1. Przeprowadź rozmowę z chatbotem. Spróbuj zdefiniować różnice pomiędzy
botem udającym człowieka (przygotowywanym na test Turinga) a botem
„asystentem, służącym”.

Bot podczas rozmowy towarzyskiej stara się imitować człowieka, może opowiedziec jakiś żart, odpowiada "luźnym" językiem. Można z nim przeprowadzić rozmowę towarzyską, ale to nie prowadzi do żadnego celu, czat nie wyrazi swoich poglądów i emocji bo ich nie ma, brakuje mu elementu ludzkiego, semiotyki.
Bot "służący" zna większość odpowiedzi na pytania, bo są to pytania naukowe, a czatboty mają wgrane odpowiednie bazy danych, z których pobierają treści, znają artykułów, książek, dlatego z nich wykorzystują informacje na podstawie których udzielają odpowiedzi na pytania. Kiedy uzywamu bota jako "służącego" rozmowa jest prosta i zrozumiała, a informacje zwrotne sątakie, jakich oczekiwaliśmy.


2. Sprawdź dwa boty z obu z tych rodzajów na występowanie zachowań:

a) opowiadanie żartów,
b) przytaczanie cytatów z twoich wypowiedzi, lub znanych osób,
c) nawiązywanie wypowiedzi do słów kluczowych,
d) zadawanie dużej liczby pytań,
e) powracanie do początku wypowiedzi, sekwencyjne powtarzanie,
f) zadawanie pytań powstających z twoich wypowiedzi,
g) odpowiadanie wymijająco, ogólnikowo,
h) częsta zmiana tematu rozmowy,
i) problemy z utrzymaniem wątków.

3. Sporządź raport ze spostrzeżeń.

czat umie opowiadać żarty, potrafi przytaczać cytaty i wypowiedzi. Na pytania naukowe odpowiada pełnymi zdaniami, odpowiada na praktycznie wszystkie zadane pytania zna odpowiedzi, nie zmienia tematów, utzymuje wątek.

4. Na podstawie powyższych obserwacji, w grupie dwóch osób spróbujcie prze-
widzieć zachowania dwóch rozmawiających ze sobą chatbotów (przepisując ich
wzajemne odpowiedzi).

W takim przypadku czatboty powtarzają w kółko te same kwestie, nie znaja odpowiedzi na niektóre pytania, odpowiadaja pytaniem na pytanie.

5. Zdenerwuj bota 😊😊

Po zbyt duzej ilości zadanych pytań czt powtarza się.
