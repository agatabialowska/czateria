```python
% zadanie 1

% Fakty

% Ojciec
ojciec(filip, anna).
ojciec(filip, karol).
ojciec(filip, andrzej).
ojciec(filip, edward).
ojciec(jerzy, elzbieta).
ojciec(jerzy, malgorzata).

ojciec(karol, wiliam).
ojciec(karol, harry).
ojciec(antony, dawid).
ojciec(antony, sara).

% Matka
matka(elzbieta, anna).
matka(elzbieta, karol).
matka(elzbieta, andrzej).
matka(elzbieta, edward).
matka(malgorzata, dawid).
matka(malgorzata, sara).
matka(diana, wiliam).
matka(diana, harry).
matka(elizabeth, elzbieta).
matka(elizabeth, malgorzata).

% Małżeństwa
malzenstwo(filip, elzbieta).
malzenstwo(karol, diana).
malzenstwo(antony, malgorzata).
malzenstwo(jerzy, elizabeth).

% Reguła małżeństwa (związek małżeński)
malzenstwo(X, Y) :-
    malzenstwo(Y, X). % Dodajemy symetrię małżeństwa

% Płeć
mezczyzna(filip).
mezczyzna(karol).
mezczyzna(andrzej).
mezczyzna(edward).
mezczyzna(jerzy).
mezczyzna(dawid).
mezczyzna(wiliam).
mezczyzna(harry).
mezczyzna(antony).

kobieta(elzbieta).
kobieta(anna).
kobieta(malgorzata).
kobieta(sara).
kobieta(diana).
kobieta(elizabeth).

% Reguły

% Reguła ogólna rodzica
rodzic(X, Y) :- ojciec(X, Y).
rodzic(X, Y) :- matka(X, Y).


% Reguła przodka: X jest przodkiem Y, jeśli X jest rodzicem Y lub przodkiem rodzica Y
przodek(X, Y) :-
    rodzic(X, Y).
przodek(X, Y) :-
    rodzic(X, Z),
    przodek(Z, Y).

% Brat(X, Y): X jest bratem Y, jeśli X i Y mają wspólnych rodziców i X jest mężczyzną
brat(X, Y) :-
    mezczyzna(X),
    rodzic(Z, X),
    rodzic(Z, Y),
    X \= Y.

% Siostra(X, Y): X jest siostrą Y, jeśli X i Y mają wspólnych rodziców i X jest kobietą
siostra(X, Y) :-
    kobieta(X),
    rodzic(Z, X),
    rodzic(Z, Y),
    X \= Y.

% Kuzyn(X, Y): X jest kuzynem Y, jeśli ich rodzice są rodzeństwem
kuzyn(X, Y) :-
    rodzic(Z, X),
    rodzic(W, Y),
    (brat(Z, W); siostra(Z, W)).  % Sprawdzamy, czy rodzice X i Y są rodzeństwem



% zapytania:
% zapytanie o przodka:
% przodek(filip, anna).
% true

% przodek(filip, harry).
% true

% zapytanie o rodzeństwo:
% siostra(elzbieta, malgorzata).
% true

% brat(wiliam, harry).
% true

% wypisanie wszystkich braci:
% setof(Brat, brat(Brat, edward), Bracia).
% Bracia = [andrzej, karol]

% wypisanie wszystkich przodków:
% findall(X, przodek(X, wiliam), Wyniki).
% Wyniki = [karol, diana, filip, jerzy, elzbieta, elizabeth]
```

```python
% zadanie 3 zadanie dziala w SWI prolog

% Deklaracja dynamicznego predykatu
:- dynamic produkt/3.

% Produkty
produkt(telefon, 1, sektor_a).
produkt(odkurzacz, 2, sektor_b).
produkt(piekarnik, 3, sektor_c).
produkt(tv, 4, sektor_d).

% Reguła sprawdzająca dostępność produktu w magazynie
dostepny_produkt(X) :-
    produkt(X, Ilosc, _),
    Ilosc > 0.

% Reguła przenosząca produkt do innego sektora
przenies_produkt(X, Y) :-
    produkt(X, Ilosc, StarySektor),
    retract(produkt(X, Ilosc, StarySektor)),   % Usuwamy produkt ze starego sektora
    assert(produkt(X, Ilosc, Y)),               % Dodajemy produkt do nowego sektora
    format('Produkt ~w został przeniesiony z ~w do ~w.~n', [X, StarySektor, Y]).

% Reguła uzupełniająca stan produktu
uzupelnij_stan(X, N) :-
    produkt(X, Ilosc, Sektor),
    retract(produkt(X, Ilosc, Sektor)),        % Usuwamy stary stan produktu
    NowaIlosc is Ilosc + N,                    % Zwiększamy ilość produktu
    assert(produkt(X, NowaIlosc, Sektor)),     % Dodajemy nowy stan produktu
    format('Produkt ~w, nowa ilość: ~w, Sektor: ~w~n', [X, NowaIlosc, Sektor]).


% Zapytania:
% zapytanie o dostepnosc?
% dostepny_produkt(tv).
% true

% dostepny_produkt(smartfon).
% false

% zapytanie o przeniesienie produktu?
% przenies_produkt(tv, sektor_b).
% Produkt tv został przeniesiony z sektor_d do sektor_b.
% true

% przenies_produkt(odkurzacz, sektor_c).
% Produkt odkurzacz został przeniesiony z sektor_b do sektor_c.
% true

% zapytanie o uzupełnienie stanów?
% uzupelnij_stan(odkurzacz, 3).
% produkt odkurzacz, nowa ilość: 5, Sektor: sektor_b
% true

% uzupelnij_stan(telefon, 11).
% Produkt telefon, nowa ilość: 12, Sektor: sektor_a
% true
```

```python
% zadanie 4

% Połączenia drogowe między miastami (czas przejazdu w godzinach)
droga(lublin, wroclaw, 9).
droga(lublin, krakow, 5).
droga(krakow, wroclaw, 4).
droga(wroclaw, poznan, 3).
droga(krakow, poznan, 6).
droga(lublin, poznan, 10).

% Reguła sprawdzająca dostępność drogi
droga(X, Y) :-
    droga(X, Y, _).  % Sprawdzenie, czy istnieje bezpośrednia droga

droga(X, Y) :-
    droga(Y, X, _).  % Dodajemy symetrię, bo droga jest dwukierunkowa

% Reguła obliczająca czas przejazdu między miastami
czas_przejazdu(X, Y, T) :-
    droga(X, Y, T).  % Bezpośrednia droga

czas_przejazdu(X, Y, T) :-
    droga(X, Z, T1), % Sprawdzamy połączenie z miastem pośrednim Z
    czas_przejazdu(Z, Y, T2),  % Rekursywnie obliczamy czas z miasta Z do Y
    T is T1 + T2.  % Całkowity czas to suma czasów T1 i T2

najkrotsza_trasa(X, Y, NajkrotszaTrasa, NajkrotszyCzas) :-
    findall((Trasa, Czas), trasa(X, Y, Trasa, Czas), MozliweTrasy),
    minimalna_trasa(MozliweTrasy, (NajkrotszaTrasa, NajkrotszyCzas)).

trasa(X, Y, [X, Y], Czas) :- droga(X, Y, Czas).
trasa(X, Y, [X | ResztaTrasy], Czas) :-
    droga(X, Z, Czas1),
    trasa(Z, Y, ResztaTrasy, Czas2),
    X \= Y,
    \+ member(X, ResztaTrasy),
    Czas is Czas1 + Czas2.
minimalna_trasa([Trasa], Trasa).
minimalna_trasa([(T1, C1), (_, C2) | Reszta], Min) :-
    C1 =< C2, minimalna_trasa([(T1, C1) | Reszta], Min).
minimalna_trasa([(_, C1), (T2, C2) | Reszta], Min) :-
    C1 > C2, minimalna_trasa([(T2, C2) | Reszta], Min).



% Zapytania:
% sprawdzenie czy jest bezpośrednia droga?
% droga(krakow, wroclaw).
% true

% droga(lublin, krakow).
% true

% zapytanie o czas przejazdu?
% czas_przejazdu(lublin, wroclaw, T).
% T = 9

% czas_przejazdu(wroclaw, poznan, T).
% T = 3

% zapytanie o najkrótszą trasę między miastami?
% najkrotsza_trasa(lublin, poznan, Trasa_przez, Czas).
% Czas = 10,
% Trasa_przez = [lublin, poznan]

% najkrotsza_trasa(wroclaw, poznan, Trasa_przez, Czas).
% Czas = 3,
% Trasa_przez = [wroclaw, poznan]
```
